```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import cv2
import os
from pathlib import Path
```

```python
traindata=pd.read_csv("train.csv")
testdata=pd.read_csv("test.csv")
labels_dict={"Pikachu":0,"Bulbasaur":1,"Charmander":2}
label1_dict={0:"Pikachu",1:"Bulbasaur",2:"Charmander"}
image_data=[] ##train image data
labels=[]  ##actual label
testimage_data=[] ##test image data
IMG_SIZE=28
```

```python
data=traindata.values ##convert to numpy array for slicing
Y=data[:,1]
X=data[:,0]
```

```python
for i in Y:
    labels.append(labels_dict[i])
```

```python
for imagename in X:
    img_array=cv2.imread("Images\\"+(imagename))
    img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB)
    img_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    #img_array=img_array[2:26,2:26,:]
    img_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    img_array=img_array/255
    image_data.append(img_array)
```

```python
image_data=np.array(image_data)
print(len(image_data))
print(len(labels))
```

```python
x=image_data
y=np.array(labels)
print(x.shape)
print(y.shape)
```

```python
for i in range(10):
    r=np.random.randint(304)
    plt.imshow(x[r])
    plt.title(y[r])
    plt.show()
```

```python
test_data=testdata.values #Forming the test data
test_data=test_data[:,0]
```

```python
for test_image in test_data:
    testimg_array=cv2.imread("Images\\"+(test_image))
    testimg_array=cv2.cvtColor(testimg_array,cv2.COLOR_BGR2RGB)
    testimg_array=cv2.resize(testimg_array,(IMG_SIZE,IMG_SIZE))
    #testimg_array=testimg_array[2:26,2:26,:]
    testimg_array=cv2.resize(testimg_array,(IMG_SIZE,IMG_SIZE))
    testimg_array=testimg_array/255
    testimage_data.append(testimg_array)
```

```python
testimage_data=np.array(testimage_data)
print(len(testimage_data))
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(IMG_SIZE,IMG_SIZE,3)),
    keras.layers.Dense(260, activation=tf.nn.relu),
    keras.layers.Dense(40, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=30)
```

```python
 predictions=model.predict(testimage_data)
```

```python
ans=[]
for i in predictions:
    r=np.argmax(i)
    ans.append(label1_dict[r])
```

```python
ans
```

```python
testdata['NameOfPokemon']=ans
df=testdata
df.to_csv('submitchallengenew112.csv',index=False)
```
