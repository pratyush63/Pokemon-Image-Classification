```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
traindata=pd.read_csv("train.csv")
testdata=pd.read_csv("test.csv")
labels_dict={"Pikachu":0,"Bulbasaur":1,"Charmander":2}
label1_dict={0:"Pikachu",1:"Bulbasaur",2:"Charmender"}
image_data=[]
labels=[]
testimage_data=[]
```

```python
data=traindata.values ##convert to numpy array for slicing
Y=data[:,1]
X=data[:,0]
```

```python
for i in Y:
    labels.append(labels_dict[i])
```

```python
IMG_SIZE=200
for imagename in X:
    img_array=cv2.imread("Images\\"+(imagename))
    img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB)
    #img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB)
    img_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    img_array=img_array[65:135,65:135,:]
    img_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    img_array=img_array/255
    image_data.append(img_array)
```

```python
image_data=np.array(image_data)
print(len(image_data))
print(len(labels))
```

```python
x=image_data
y=np.array(labels)
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5
```

```python
def knn(X,Y,test_point,k=5):
    m=X.shape[0]
    vals=[]
    for i in range(m):
        d=dist(test_point,X[i])
        vals.append((d,Y[i]))
        
    vals=sorted(vals)
    vals=np.array(vals[:k])
    b=np.unique(vals[:,1],return_counts=True)
    idx=np.argmax(b[1])
    pred=b[0][idx]
    return int(pred)
```

```python
test_data=testdata.values
test_data=test_data[:,0]
```

```python
for test_image in test_data:
    testimg_array=cv2.imread("Images\\"+(test_image))
    testimg_array=cv2.cvtColor(testimg_array,cv2.COLOR_BGR2RGB)
    #testimg_array=cv2.cvtColor(testimg_array,cv2.COLOR_BGR2RGB)
    testimg_array=cv2.resize(testimg_array,(IMG_SIZE,IMG_SIZE))
    testimg_array=testimg_array[65:135,65:135,:]
    testimg_array=cv2.resize(testimg_array,(IMG_SIZE,IMG_SIZE))
    testimg_array=testimg_array/255
    testimage_data.append(testimg_array)
```

```python
testimage_data=np.array(testimage_data)
print(len(testimage_data))
```

```python
ans=[]
for tm in testimage_data:
    r=knn(x,y,tm,k=1)
    ans.append(label1_dict[r])
```
